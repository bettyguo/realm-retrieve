A 3-minute walkthrough of the **ReaLM-Retrieve** pipeline using its CPU-friendly toy components:

1. Segment a reasoning chain into steps.
2. Score each step with **RSUS** (Reasoning Step Uncertainty Score).
3. Decide whether to **retrieve** with a policy threshold.
4. Retrieve with **BM25** (toy stand-in for ColBERTv2 + PLAID).
5. Extract a final answer and report metrics.

Swap the toy components for the production ones (`ColBERTRetriever`, `VLLMReasoningModel`, …) when you're ready to scale.

## 1 — Build a toy corpus and retriever

```python
from realm_retrieve.toy import (
    ToyPipeline, ToyReasoningModel, ToyRetriever,
    demo_corpus, demo_questions,
)

corpus = demo_corpus()
retriever = ToyRetriever(corpus, k=3)
print(f'Corpus: {retriever.get_corpus_size()} docs')
for d in corpus[:3]:
    print(f'  [{d.doc_id}] {d.text[:80]}...')
```

## 2 — Inspect a single question

```python
lrm = ToyReasoningModel(seed=42)
pipeline = ToyPipeline(retriever, lrm, rsus_threshold=0.5)

q, gold = 'What is the capital of Sweden?', 'Stockholm'
result = pipeline.answer(q, gold=gold)
print(result)
```
